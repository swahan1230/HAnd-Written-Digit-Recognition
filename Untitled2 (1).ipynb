```python
def show_imgs(imgs,title):
  img_grid=torchvision.utils.make_grid(imgs,nrow=20)
  img_np=img_grid.numpy()
  img_np=np.transpose(img_np,(1,2,0))
  plt.figure(figsize=(100,100))
  plt.imshow(img_np)
  plt.title(title)
  plt.show()
```

```python
class CNN_Model(nn.Module):
    def __init__(self):
        super(CNN_Model,self).__init__()
        self.cnn1 = nn.Conv2d(in_channels=1, out_channels=16, kernel_size=5, stride=1, padding=0)
        self.relu1 = nn.ReLU()
        
        self.maxpool1 = nn.MaxPool2d(kernel_size=2)
     
        self.cnn2 = nn.Conv2d(in_channels=16, out_channels=32, kernel_size=5, stride=1, padding=0)
        self.relu2 = nn.ReLU()
        
        self.maxpool2 = nn.MaxPool2d(kernel_size=2)
        
        self.fc1 = nn.Linear(32 * 4 * 4, 10)
    def forward(self, x):
        # Convolution 1
        out = self.cnn1(x)
        out = self.relu1(out)
        
        # Max pool 1
        out = self.maxpool1(out)
        
        # Convolution 2 
        out = self.cnn2(out)
        out = self.relu2(out)
        
        # Max pool 2 
        out = self.maxpool2(out)
        
        # flatten
        out = out.view(out.size(0), -1)

        # Linear function (readout)
        out = self.fc1(out)
        
        return out
```

```python
def accuracy(model, dataloader):
    correct, total = 0, 0
    model.eval()
    with torch.no_grad():
        for images, labels in dataloader:
            images, labels = images.to(device), labels.to(device)
            pred_probs = model(images)
            pred_labels = torch.argmax(pred_probs, 1)

            total += len(labels)
            correct += (pred_labels == labels).sum().item()

    return correct / total * 100

# Training
%reload_ext tensorboard
%tensorboard --logdir runs

#model = CNN_Model()
#model=model.to(device)
loss_fn = nn.CrossEntropyLoss()
adam = optim.Adam(model.parameters(),lr=0.001)

epochs = 10
writer = SummaryWriter('runs/mnist-exp1')
step_no = 0

min_loss= 1000

for epoch in tqdm(range(0, epochs), total=epochs, desc='Training'):
    running_loss = []
    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)
        # FP
        model.train()
        preds = model(images)
        loss = loss_fn(preds, labels)

        # logging loss values
        writer.add_scalar('Step loss', loss.item(), step_no)
        running_loss.append(loss.item())
        
        # BP
        adam.zero_grad()
        loss.backward()
        adam.step()
        step_no += 1

        del images,labels,preds
        torch.cuda.empty_cache()

    # mean of all step losses = epoch loss
    epoch_loss = sum(running_loss) / len(running_loss)
    train_acc = accuracy(model, train_loader)
    test_acc = accuracy(model, test_loader) 
    
    #model checkpointing
    if min_loss > epoch_loss:
      min_loss= epoch_loss
      torch.save(model, 'model2.h5')

    writer.add_scalar('Epoch loss', epoch_loss, epoch)
    writer.add_scalars('Accuracies', {
        'Train_acc': train_acc,
        'Test_acc': test_acc
    }, epoch)

writer.close()
```
